```python
import requests

from bs4 import BeautifulSoup

page = requests.get("https://en.wikipedia.org/wiki/2024_Summer_Olympics_medal_table")

page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
soup1 = soup.prettify()
soup1
```

```python
table = soup.find("table",class_ = "wikitable sortable sticky-header-multi plainrowheaders jquery-tablesorter").tbody
```

```python
rows = table.find_all('tr')
cols = [val.text.replace('\n','')for val in rows[0].find_all('th')]
cols
```

```python
row_headers = []

for i in range(1,len(rows)):
    row_headers.append([val.text.replace('\n','') for val in rows[i].find_all('th')])
row_headers
```

```python
scores = []

for i in range(1,len(rows)):
  scores.append([val.text.replace('\n','') for val in rows[i].find_all('td')])
```

```python
scores
```

```python
scores
```

```python
for i in range(0,len(scores)):
  if len(scores[i]) < 5:
    scores[i].insert(0,0)

scores
```

```python
ranks = []

for i in range(0,len(scores)):
  if scores[i]:
    ranks.append([scores[i][0]])
ranks
```

```python
data = []

for i in range(0,len(scores)):
  data.append(row_headers[i]+scores[i])
data
```

```python
import pandas as pd

df = pd.DataFrame(data,columns=cols)

df
```

```python
df = df.rename(columns = {'Rank':'NOC','NOC':'Ranks'})
```
