```python
import pandas as pd

data = pd.read_csv('/content/cricket.csv')

data.head()
```

```python
data.duplicated()
```

```python
data[data['Player'].duplicated()==1]
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.isnull().any()
```

```python
data[data['BF'].isna()==1]
```

```python
data['BF'].fillna(0,inplace=True)
```

```python
data.isnull().any()
```

```python
data.head()
```

```python
data['Span'].str.split(pat='-')
```

```python
data['Rookie_Year']= data['Span'].str.split(pat='-').str[0]
data['Final_Year'] = data['Span'].str.split(pat='-').str[1]
data.drop('Span',axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Player'].str.split(pat='(')
```

```python
data['Country'] = data['Player'].str.split('(').str[1]
data['Country'] = data['Country'].str.split(')').str[0]
data['Player'] = data['Player'].str.split('(').str[0]
```

```python
data.head()
```

```python
data.rename(columns = {'Mat':'Matches','NO':'No_Outs','HS':'Highest_Inns_Score','Ave':'Average','BF':'Balls_Faced','SR':'Batting_Strike_Rate'},inplace = True)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['Highest_Inns_Score'] = data['Highest_Inns_Score'].str.split(pat = '*').str[0]
```

```python
data['Highest_Inns_Score'] = data['Highest_Inns_Score'].astype('int')
```

```python
data.dtypes
```

```python
data = data.astype({'Rookie_Year':'int','Final_Year':'int'})
```

```python
data.head()
```

```python
data['Batting_Strike_Rate'] = data['Batting_Strike_Rate'].astype('int')
```

```python
data.dtypes
```

```python
data['Matches'] = data['Matches'].str.split(pat = '*').str[0]
```

```python
data['Matches'] = data['Matches'].astype('int')
```

```python
data.dtypes
```

```python
data['Balls_Faced'] = data['Balls_Faced'].str.split(pat = '+').str[0]
```

```python
data['Balls_Faced'].isna()==1
```

```python
data['Balls_Faced'].fillna(0,inplace=True)
```

```python
data['Balls_Faced'] = data['Balls_Faced'].astype('int')
```

```python
data['Average'] = data['Average'].astype('int')
```

```python
data['4s'] = data['4s'].str.split(pat = '+').str[0]
```

```python
data['4s'] = data['4s'].astype('int')
```

```python
data['6s'] = data['6s'].str.split(pat = '+').str[0]
```

```python
data['6s'] = data['6s'].astype('int')
```

```python
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import Figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#Correlation between Matches vs Runs(This can show how consistent players score with more opportunities.)
```

```python
correlation_matrix = data.corr(method = 'pearson',numeric_only = True)
sns.heatmap(correlation_matrix, annot = True)
plt.show()
```

```python
correlation = data['Matches'].corr(data['Runs'])
plt.scatter(data['Matches'],data['Runs'],color = 'blue')
plt.title('Matches vs Runs')
plt.xlabel('Matches')
plt.ylabel('Runs')
plt.show()
```
