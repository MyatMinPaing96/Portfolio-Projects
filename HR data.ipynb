**Correlation Between Hire Date Vs Salary**

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
from matplotlib.pyplot import Figure
plt.rcParams['figure.figsize'] = (8,6)
```

```python
data.dtypes
```

```python
data['hire_date'] = pd.to_datetime(data['hire_date'])
```

```python
data['hire_date'].dtypes
```

```python
data['hire_date_ordinal'] = data['hire_date'].apply(lambda date: date.toordinal())
data['hire_date_ordinal']
```

```python
data.head()
```

```python
correlation = data['hire_date_ordinal'].corr(data['salary'])
```

```python
plt.bar(data['hire_date_ordinal'],data['salary'])
plt.xlabel('Hire Date')
plt.ylabel('Salary')
plt.title('Hire Date Vs Salary')
plt.show()
```

```python
data['height'] = data['height']/data['height'].max()

data.head()
```

```python
data['weight'] = (data['weight']-data['weight'].min())/(data['weight'].max()-data['weight'].min())

data.head()
```

```python
data['salary'] = (data['salary']-data['salary'].mean())/data['salary'].std()

data.head()
```

**Show the salary with the Salary Group.**

```python
import numpy as np

bins = np.linspace(min(data['salary']),max(data['salary']),6)
bins
```

```python
bin_names = ['Very Low','Low','Medium','High','Very High']


data['salary_group'] = pd.cut(data['salary'],bins,labels=bin_names,include_lowest=True)

data.head()
```
