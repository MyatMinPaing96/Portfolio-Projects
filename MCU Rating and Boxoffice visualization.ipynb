```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Box office gross Worldwide'],bins = 20 , kde = True)
plt.title('Distribution of Box Office Gross Worldwide')
plt.xlabel('Box Office Gross(in Billions)')
plt.ylabel('Frequency')
plt.show
```

```python
Top_Film = df[['Film','Box office gross Worldwide']].sort_values(by = 'Box office gross Worldwide',ascending = False).head(10)
Top_Film.head()
```

```python
plt.figure (figsize = (10,6))
sns.barplot(x = 'Box office gross Worldwide',y = 'Film', hue='Film',data = Top_Film, palette = 'viridis')
plt.title('Top 10 Grossing Films Worldwide')
plt.xlabel('Box Office Gross (in Billions)')
plt.ylabel('Film')
plt.show()
```

```python
df['Box office gross U.S. and Canada'] = df['Box office gross U.S. and Canada']/1e6
df['Box office gross Other territories'] = df['Box office gross Other territories']/1e6
Film_Top_10 = df[['Film','Box office gross U.S. and Canada','Box office gross Other territories']].sort_values(by = 'Box office gross U.S. and Canada',ascending = False).head(10)

Film_Top_10.set_index('Film').plot(kind='barh', stacked=False, figsize=(12, 7))
plt.title('Top 10 Grossing Films Worldwide')
plt.xlabel('Box Office Gross (in Millions)')
plt.ylabel('Film')
plt.show()
```

```python
Rating = df[['Film','Rotten Tomatoes Rating']].sort_values(by = 'Rotten Tomatoes Rating',ascending=False)
plt.figure(figsize = (10,6))
sns.barplot(x = 'Rotten Tomatoes Rating',y = 'Film', hue='Film',data = Rating, palette = 'viridis')
```
