```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import Figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
data = pd.read_csv('/content/movies.csv')

data.head(5)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

```python
for col in data.columns:
  pct_missing = np.mean(data[col].isnull())
  print('{} - {}%'.format(col, pct_missing))
```

```python
data.dtypes
```

```python
#change data types of columns

data['budget'] = data['budget'].astype('int64')
data['gross'] = data['gross'].astype('int64')
```

```python
data.head()
```

```python
data = data.sort_values(by=['gross'],inplace = False ,ascending= False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
#Budget high correlation
#Company high corelation

```

```python
plt.scatter(x = data['budget'],y = data['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross for Film')
plt.show()
```

```python
data.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = data, scatter_kws = {'color': 'red'},line_kws= {'color':'blue'})
```

```python
data['name'] = data['name'].astype(str)
```

```python
#Correlation

data.corr(method ='pearson' , numeric_only=True)
```

```python
correlation_matrix = data.corr(method = 'pearson',numeric_only = True)

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
