```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('/content/amazon.csv')
df.head(3)
```

```python
df.info()
```

```python

df.isnull().sum()
```

```python
df[df['rating_count'].isnull() == 1]
```

```python
df['rating_count'] = df['rating_count'].str.replace(',','').astype('float')
```

```python
df['rating_count'] = df.rating_count.fillna(value=df['rating_count'].median())
```

```python
df['rating'] = df['rating'].str.replace('|','3.9').astype('float')
```

```python
df.duplicated().sum()
```

```python
df['discounted_price'] = df['discounted_price'].astype('str')
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ','',regex = True).str.replace(',','',regex = True)
```

```python
df['discounted_price'] = df['discounted_price'].astype('float')
```

```python
df['actual_price'] = df['actual_price'].astype('str')
df['actual_price'] = df['actual_price'].str.replace('‚Çπ','',regex = True).str.replace(',','',regex = True)
df['actual_price'] = df['actual_price'].astype('float')
```

```python
df['discount_percentage'] = df['discount_percentage'].astype('str')
df['discount_percentage'] = df['discount_percentage'].str.replace('%','',regex = True)
df['discount_percentage'] = df['discount_percentage'].astype('int')
```

```python
df.head(2)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df.info()
```

```python
df[['Main Category','category1','category2']] = df['category'].astype('str').str.split('|',expand = True,n=2)
```

```python
discount_by_category = df.groupby('Main Category',observed = False)['discount_percentage'].mean()
discount_by_category
```

```python
plt.figure(figsize = (10,6))
plt.barh(discount_by_category.index,discount_by_category.values)
plt.xlabel('Discount Percentage')
plt.ylabel('Main Category')
plt.title('Discount Percentage per Category')
plt.show()
```

```python
df['rating'].value_counts()
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = 'actual_price', y = 'rating',data = df)
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = 'discounted_price', y = 'rating',data = df)
```

```python
correlation1 = df['discount_percentage'].corr(df['rating'])
print(f"The Correlation between discount_percentage and rating is : {correlation1}")
```

```python
correlation_matrix = df.corr(numeric_only = True)
correlation_matrix
```

```python
sns.heatmap(correlation_matrix,annot = True,cmap = 'coolwarm')
```
