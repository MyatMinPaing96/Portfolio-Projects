```python
paris_listing_neighbourhood = newdata.groupby('neighbourhood')['price'].mean().sort_values(ascending=True)
paris_listing_neighbourhood
```

```python
paris_listing_accommodates = newdata.query("neighbourhood == 'Elysee'").groupby('accommodates').agg({'price': 'mean'}).sort_values('price')
paris_listing_accommodates
```

```python
paris_listings_over_time = newdata.set_index('host_since').resample("YE").agg({'neighbourhood':'count','price': 'mean'})
paris_listings_over_time
```

```python
import matplotlib.pyplot as plt

paris_listing_neighbourhood.plot(kind='barh')
plt.xlabel('Average Price per night(Euros)')
plt.ylabel('Neighbourhood')
plt.title('Average Listing Price by Neighbourhood')
plt.show()
```

```python
paris_listing_accommodates.plot(kind='barh',color = 'green')
plt.xlabel('Average Price per night(Euros)')
plt.ylabel('Accommodates Capacity')
plt.title('Average Listing Price by Accommodates Capacity')
plt.show()
```

```python
paris_listings_over_time["neighbourhood"].plot(ylabel='New Host',title = 'New Airbnb host in Paris Over Time')
plt.show()
```

```python
paris_listings_over_time["price"].plot(ylabel='Average Price',title = 'Average Airbnb Price in Paris Over Time')
plt.show()
```

```python
fig,ax = plt.subplots()

ax.plot(
    paris_listings_over_time.index,
    paris_listings_over_time['neighbourhood'],
    label = 'New Host',
    color = 'teal'
)
ax.set_ylabel('New Host')

ax2 = ax.twinx()
ax2.plot(
    paris_listings_over_time.index,
    paris_listings_over_time['price'],
    label = 'Average Price',
    color = 'Orange'
)
ax2.set_ylim(0)
ax2.set_ylabel('Average Price')

ax.title.set_text('New Airbnb Host and Average Airbnb Price in Paris Over Time')
plt.show()
```
